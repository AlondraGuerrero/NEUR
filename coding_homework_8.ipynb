```python
# Import modules
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Import files from shared drive and github
url1 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/extinction_predictors_hw8.csv'
url2 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/fear_conditioning_traces_hw8.csv'
fear_traces = np.loadtxt(url2, delimiter = ',')
extinction_predictors = np.loadtxt(url1, delimiter =',')
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/shock_index.csv', delimiter = ',').astype(int)
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/tone_index.csv', delimiter = ',').astype(int)
```

```python
# Make two new variables - conditioning_traces and extinction_traces
conditioning_traces= fear_traces[0:10458,:]
extinction_traces= fear_traces[10458:,:]
print(conditioning_traces)
print(extinction_traces)
```

```python
# Print the size of conditioning_traces and extinction_traces arrays
print(conditioning_traces.size)
print(extinction_traces.size)
```

```python
# Make time_conditioning and time_extinction variables
time_conditioning = np.linspace(0,697.2,10458)
time_extinction = np.linspace(0,947.3,14210)
```

```python
#
fig, ax = plt.subplots(2,figsize=(15,5 ))
ax[0].plot(time_conditioning, conditioning_traces[:,0], color = "blue")
ax[0].plot(time_conditioning, conditioning_traces[:,1], color = "red")
ax[1].plot(time_extinction, extinction_traces[:,0], color = "blue")
ax[1].plot(time_extinction, extinction_traces[:,1], color = "red")

ax[0].vlines(shock_index, 0,1, 'k', lw=2)
ax[0].vlines(tone_index, 0,1, 'g', lw=2)

ax[0].set_xlabel('time')
ax[0].set_ylabel('traces')
ax[1].set_xlabel('time')
ax[1].set_ylabel('traces')
```

```python
# Make a predictors_conditioning variable that contains shock onset and tone onset
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params
print('b1:')
print(b1)
```

```python
# Create the r_model_conditioning variable, and run for loop
predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r3.statistic
```

```python
# Runs another loop that does the same thing for extinction_traces array.
shock_array = np.zeros(14210)
for i in range(14210):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14210)
for i in range(14210):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = extinction_traces[:,0]

predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
model3 = sm.GLM(first_trace, predictors_extinction, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params
print('b1:')
print(b1)

predictors_extinction['tone onset'] = tone_array
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r3 = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r3.statistic
```

```python
# Use the pearsonr function to correlate r_index_conditioning and r_index_extinction variables
pearsonr(r_model_conditioning, r_model_extinction)
```
