```python
# Imports modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Imports files from shared drive folder
url1 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/acc_hits_hw6.csv'
url2 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/acc_fa_hw6.csv'
url3 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/lc_hits_hw6.csv'
url4 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/lc_fa_hw6.csv'
acc_hits = np.loadtxt(url1,delimiter=",")
acc_fa = np.loadtxt(url2,delimiter=",")
lc_hits = np.loadtxt(url3,delimiter=",")
lc_fa = np.loadtxt(url4,delimiter=",")
```

```python
# Prints shape of each array 
print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
# Creates variable called time
time = np.linspace(0,4,8001)
```

```python
# Expression to determine sampling rate
8001/4
```

```python
# Plots the first trial of acc_hits and lc_hits on one subplot and acc_fa and lc_fa in another subplot
fig, ax = plt.subplots(2,figsize=(9,6 ))
ax[0].plot(time, acc_hits[0], color = "red")
ax[0].plot(time, lc_hits[0], color = "blue")
ax[0].set_ylabel('Voltage (mV)')
ax[0].set_xlabel('time')
ax[1].plot(time, acc_fa[0], color = "red")
ax[1].plot(time, lc_fa[0], color = "blue")
ax[1].set_ylabel('Voltage (mV)')
ax[1].set_xlabel('time')
```

The dominat oscillation for ACC LFP is around 2.5 while the dominat oscillation for LC LFP is around 2.0, meaning they are different. There does seem to be a relationshp in the raw LFP where sometimes the peaks and troughs of two oscillations have a consistent relationship across time. 

```python
# Compute the coherence for acc_hits and lc_hits & it defines it in order to call it later
def coherence1():
  dt = time[1] - time[0]              
  K = acc_hits.shape[0]              
  N = lc_hits.shape[1]  
  xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  
  yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits]) 

  T = time[-1]
  scale = 2 * dt**2 / T
  Sxx = scale * (xf * xf.conj()).mean(0)  
  Syy = scale * (yf * yf.conj()).mean(0)  
  Sxy = scale * (xf * yf.conj()).mean(0)  

  cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

  f = np.fft.rfftfreq(N, dt)                     
  plt.plot(f, cohr.real, color = "red")                      
  plt.xlim([0, 50])                           
  plt.ylim([0, 1])                            
  plt.xlabel('Frequency [Hz]')             
  plt.ylabel('Coherence')

  plt.show()
plt.title('Coherence between acc_hits and lc_hits')
coherence1()

```

The frequency band where there is the most coherence would be at around 7-8 Hz since that is the peak that is closest to 1 (1 indicates strong coherence between signals).

```python
# Compute the coherence for acc_fa and lc_fa
dt = time[1] - time[0]              
K = acc_fa.shape[0]              
N = lc_fa.shape[1]  
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa]) 

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = "blue")                     
plt.xlim([0, 50])                           
plt.ylim([0, 1])                            
plt.xlabel('Frequency [Hz]')             
plt.ylabel('Coherence')
plt.title('Coherence for hits and false alarms')

# Plots coherence between acc_hits and lc_hits
coherence1()

```

The frequency at which coherence is strongest is different between hits and false alarms since false alarm's strongest coherence is around 1-2 Hz. Although it is hard to tell, coherence is higher during hits than in false alarms. Since higher coherence means more communication, the two brain regions are talking to each other more during the hits than the false alarms.
