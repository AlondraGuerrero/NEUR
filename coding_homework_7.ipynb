```python
# Import important modules
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Imports files from shared drive folder
url1 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/spikes_left_trial1.csv'
url2 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/spikes_left_trial2.csv'
url3 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/spikes_right_trial1.csv'
url4 = 'https://raw.githubusercontent.com/AlondraGuerrero/NEUR-265/main/spikes_right_trial2.csv'
spikes_left_trial1 = np.loadtxt(url1,delimiter=",")
spikes_left_trial2 = np.loadtxt(url2,delimiter=",")
spikes_right_trial1 = np.loadtxt(url3,delimiter=",")
spikes_right_trial2 = np.loadtxt(url4,delimiter=",")
```

```python
# Prints lenght of arrays
print(len(spikes_left_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial1))
print(len(spikes_right_trial2))
```

```python
# Expression to determine firing rate
print((65+53)/30) # Left (trial 1&2)
print((45+69)/30) # Right (trial 1&2))
```

These neurons fire more before they turn left since the firing rate is 3.933 compared to right's firing rate which 3.8. 

```python
# Plot the spike train data from trial 1 in your first subplot, and the spike train data from trial 2 in your second subplot.
fig, ax = plt.subplots(2,figsize=(10,6 ))
ax[0].plot(spikes_left_trial1, np.ones_like(spikes_left_trial1), '.', color = 'red')  
ax[0].plot(spikes_right_trial1, 2 * np.ones_like(spikes_right_trial1), '.') 
ax[0].set_xlim([0, 20])  
ax[0].set_ylim([0, 3])  
ax[0].set_xlabel('Time [s]')  
ax[1].plot(spikes_left_trial2, np.ones_like(spikes_left_trial2), '.', color = "red")  
ax[1].plot(spikes_right_trial2, 2 * np.ones_like(spikes_right_trial2), '.')  
ax[1].set_xlim([0, 20])              
ax[1].set_ylim([0, 3])              
ax[1].set_xlabel('Time [s]')                          
```

This neuron seems more "burst" during left turn trials (shown in red) since there is a big gap between firirngs while the neuron seems more regular-firing during right turn trials since the firing is more consistent and happens more requently. 

```python
# calculates inter-spike interval (ISI) arrays for each of the four spike arrays
ISIsLeft1 = np.diff(spikes_left_trial1)    
ISIsRight1 = np.diff(spikes_right_trial1) 
ISIsLeft2 = np.diff(spikes_left_trial2)     
ISIsRight2 = np.diff(spikes_right_trial2)  
```

```python
# Make four histograms - one for each of your four ISI arrays
bins = np.arange(0, 2, 0.01)
fig, ax = plt.subplots(2,2,figsize=(10,8 )) 
ax[0,0].hist(ISIsLeft1, bins, color = "red")     # Left in REd
ax[0,0].set_xlim([0, 0.25])         
ax[0,0].set_xlabel('ISI [s]')         
ax[0,0].set_ylabel('Counts')
   
ax[0,1].hist(ISIsRight1, bins)                  # Right in Blue 
ax[0,1].set_xlim([0, 0.25]) 
ax[0,1].set_xlabel('ISI [s]')         
ax[0,1].set_ylabel('Counts')
   
ax[1,0].hist(ISIsLeft2, bins, color = "red")
ax[1,0].set_xlim([0, 0.25]) 
ax[1,0].set_xlabel('ISI [s]')         
ax[1,0].set_ylabel('Counts')
   
ax[1,1].hist(ISIsRight2, bins) 
ax[1,1].set_xlim([0, 0.25]) 
ax[1,1].set_xlabel('ISI [s]')         
ax[1,1].set_ylabel('Counts')               
    
```

This neuron tends to fire in "bursts" and it seems to have higher "burstiness" in left turn trials (shown in red) since the counts are higher when they are grouped together. 

```python
# Calculate the firing rate across time for each of the four arrays and creates two subplots for the firing rate over time for each trial in each of the two subplots. 
binsize = 1                                                                   
fr_bins = np.arange(0,30,binsize)                                             
fr_Left1 = np.zeros(len(fr_bins)-1)                                                                            
for i in range(len(fr_bins)-1):                                               
  spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           
  true_values = spikes_left_trial1[spk_index]                                          
  fr_Left1[i] = (len(true_values))/binsize  

fr_Left2 = np.zeros(len(fr_bins)-1) 
for i in range(len(fr_bins)-1):                                               
  spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           
  true_values = spikes_left_trial2[spk_index]                                          
  fr_Left2[i] = (len(true_values))/binsize  

fr_Right1 = np.zeros(len(fr_bins)-1) 
for i in range(len(fr_bins)-1):                                               
  spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           
  true_values = spikes_right_trial1[spk_index]                                          
  fr_Right1[i] = (len(true_values))/binsize  

fr_Right2 = np.zeros(len(fr_bins)-1) 
for i in range(len(fr_bins)-1):                                               
  spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           
  true_values = spikes_right_trial2[spk_index]                                          
  fr_Right2[i] = (len(true_values))/binsize  

fig, ax = plt.subplots(2,figsize=(10,6 ))
ax[0].plot(fr_bins[:-1], fr_Left1, color = "red")
ax[0].plot(fr_bins[:-1], fr_Right1, color = "blue")
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')
ax[1].plot(fr_bins[:-1], fr_Left2, color = "red")
ax[1].plot(fr_bins[:-1], fr_Right2, color = "blue")
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')
```

The firing rate between left and right turn does differ at specific timepoints during the 30 seconds such as 0-5 seconds and 10-15 seconds, among others. This is not different for trial 1 and 2 in the way that there are points where the pattern differs, what does change is that for trial 1, left turns have higher firing rates while for trial 2, right turns have higher firirng rates at those points.
